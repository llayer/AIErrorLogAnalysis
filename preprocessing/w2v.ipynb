## 5. Apply word2vec

```python
reload(w2v_analysis)
model = w2v_analysis.run_word2vec(tokens['error_msg_tokenized'])
```

```python
word2index = {token: token_index for token_index, token in enumerate(model.wv.index2word)}
```

```python
model.save("data/word2vec_extended.model")
embedding_matrix = np.zeros((len(word2index)+1, 50))
for word, i in word2index.items():
    embedding_matrix[i] = model.wv[word]
np.save('data/embedding_matrix.npy', embedding_matrix)
```

```python
def encode(tokens):
    return [word2index[w] for w in tokens]
```

```python
tokens['error_msg_tokenized'] = tokens['error_msg_tokenized'].apply(encode)
```
